```python
#[1] Using map() to Find Squares of Numbers

# Program to find squares using map()
numbers = [1,2,3,4,5]
squares = list(map(lambda x: x*x, numbers))

print("Original list: ",numbers)
print("Squares list: ", squares)
```

```python
#Using filter() to Find Even Numbers

# Program to filter even numbers
numbers = [10,20,15,25,40,30,35]
even_numbers = list(filter(lambda x: x%2==0, numbers))

print("Original list: ", numbers)
print("Even numbers: ", even_numbers)
```

```python
#Using reduce() to Find Sum of Elements

# Program to find sum using reduce() 
from functools import reduce
numbers = [1, 2, 3, 4, 5]

sum_result = reduce(lambda x, y: x + y, numbers)

print("List Elements:", numbers)
print("Sum of Elements:", sum_result)


```

```python
#Function Returning Another Function
# Program demonstrating higher-order function

def multiplier(n):
    return lambda x: x * n

double = multiplier(2)
triple = multiplier(3)

print("Double of 5:", double(5))
print("Triple of 5:", triple(5))
```

```python
#Using sorted() with Lambda Function
# Program to sort tuples using lambda

students = [
    ("Anu", 88),
    ("Rahul", 70),
    ("Priya", 99),
    ("Kiran", 88)
]

sorted_students = sorted(students, key=lambda x: x[1])

print("Students Sorted by Marks:")
for student in sorted_students:
    print(student)
```

```python
#Using map() to Convert Strings to Uppercase

# Program to convert strings to uppercase using map()

names = ["virat", "bhuvii", "rajat", "padikal"]

uppercase_names = list(map(lambda x: x.upper(), names))

print("Original Names:", names)
print("Uppercase Names:", uppercase_names)
```

```python
#Using filter() to Find Positive Numbers

# Program to filter positive numbers

numbers = [-10, 20, -30, 40, 50, -60]

positive_numbers = list(filter(lambda x: x > 0, numbers))

print("Original List:", numbers)
print("Positive Numbers:", positive_numbers)
```
