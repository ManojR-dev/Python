```python
"""Q3. nums = [10, 5, 20, 8, 15]
Perform:
Find maximum value
Find minimum value
Find sum of elements
Sort the list (ascending & descending)
Find index of element 8"""

nums=[10,5,20,8,15]
print(max(nums))
print(min(nums))
print(sum(nums))
print(sorted(nums))
descending=sorted(nums, reverse=True)
print(descending)
print(nums.index(8))
```

```python
"""Q5. Problem Solving
Take a list of numbers from user input
Print:
Even numbers
Odd numbers
Largest and smallest number
Bonus:Remove duplicates from the list"""

num=(input("Enter a list"))
numbers = [int(x) for x in num.split()]
print(numbers)



even=[num for num in numbers if num % 2 == 0]
odd=[num for num in numbers if num % 2 != 0]

print(f"Even list: ", even)
print(f"Odd list: ", odd)

print("Largest number: ", max(numbers))
print("Smallest number: ", min(numbers))

```
