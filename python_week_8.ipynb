```python
#Q9. Factorial using Function
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
num=int(input("Enter a number: "))
print("\nFactorial:", factorial(num))
```

```python
#Q10. Prime Number Checker
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("\nIs Prime:", is_prime(7))
```

```python
#Q14. Voting Eligibility Checker
ages = (15, 18, 22, 16, 30)

def check_voting(age):
    if age >= 18:
        return "Eligible"
    return "Not Eligible"

print("\nVoting Eligibility:")
for age in ages:
    print(age, "-", check_voting(age))
```

```python
#Q15. Word Frequency Counter
sentence = input("\nEnter sentence: ")
words = sentence.split()

freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print("Frequency:", freq)
```

```python
#Q16. Library Book System
books = []

def add_book(book):
    books.append(book)

def remove_book(book):
    if book in books:
        books.remove(book)

def search_book(book):
    return book in books

def display_books():
    print(books)

add_book("Python")
add_book("Java")
remove_book("Java")

print("\nLibrary Books:")
display_books()
```

```python
#Q17. Student Result Analyzer
marks = [45, 78, 90, 35, 60]

def average(m):
    return sum(m) / len(m)

def passed_students(m):
    count = 0
    for i in m:
        if i > 40:
            count += 1
    return count

print("\nAverage:", average(marks))
print("Highest:", max(marks))
print("Lowest:", min(marks))
print("Passed Students:", passed_students(marks))
```

```python
#Q18. Password Strength Checker
def check_password(password):
    if (len(password) >= 8 and
        any(c.isdigit() for c in password) and
        any(c.isupper() for c in password) and
        any(c.islower() for c in password)):
        return "Strong Password"
    return "Weak Password"

pwd = input("\nEnter password: ")
print( check_password(pwd))
```

```python
#Q20. Mini Banking System
balance = 1000

def deposit(amount):
    global balance
    balance += amount

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
    else:
        print("Insufficient Balance")

def check_balance():
    print("Balance:", balance)

deposit(500)
withdraw(200)

print("\nBanking System:")
check_balance()
```

```python
#Q21. HackerRank Style Problem
nums = [10, 20, 30, 40, 50]

maximum = nums[0]
for i in nums:
    if i > maximum:
        maximum = i

minimum = nums[0]
for i in nums:
    if i < minimum:
        minimum = i

reversed_list = []
for i in range(len(nums)-1, -1, -1):
    reversed_list.append(nums[i])

print("\nMaximum:", maximum)
print("Minimum:", minimum)
print("Reversed List:", reversed_list)
```
